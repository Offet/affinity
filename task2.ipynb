```python
# Find where the truck is going and the date it is 
"""
Currently, we have 26934219.93kg to Accra, 5706381.50kg ti Bolgatanga, 
1914823.90kg to Kintampo, 3617604.59kg to Koforidua and 15919032.90kg to Tema

There are 30 Trucks, some truck have volumes of 20000kg, 13000kg, 15000, and 12000

How can we assiggn the trucks such that all 30 trucks can be equally spread to each location??

write a code to assign trucks based on location
"""

```

```python
# Sample function to assign trucks per location.
def assign_trucks(df, trucks_per_location=6):
    """
    Assigns truck IDs to order batches for each location based on date.
    
    Parameters:
        df (pd.DataFrame): DataFrame with columns: 'location', 'dateGenerated', 'volume(kg)_y'
        trucks_per_location (int): Number of trucks to assign per location.
        
    Returns:
        pd.DataFrame: DataFrame with an added 'truck_id' column indicating assignment.
    """
    assignments = []
    
    # Process each location separately.
    for location, group in df.groupby('location'):
        # Sort orders by date.
        group = group.sort_values('dateGenerated').reset_index(drop=True)
        total_weight = group['volume(kg)_y'].sum()
        target_weight = total_weight / trucks_per_location
        
        truck_id = 1
        current_weight = 0
        
        for idx, row in group.iterrows():
            # If adding the current row exceeds the target and we haven't assigned all trucks, move to next truck.
            if current_weight + row['volume(kg)_y'] > target_weight and truck_id < trucks_per_location:
                truck_id += 1
                current_weight = 0
            
            assignments.append({
                'location': location,
                'dateGenerated': row['dateGenerated'],
                'volume(kg)_y': row['volume(kg)_y'],
                'truck_id': truck_id
            })
            current_weight += row['volume(kg)_y']
    
    return pd.DataFrame(assignments)




```
